```python
import pandas as pd 
import numpy as np
```

```python
movies= pd.read_csv("movies.csv")
movies.head()
```

```python
ratings= pd.read_csv("ratings.csv")
```

```python
ratings.head()
```

```python
movies.shape
```

```python
ratings.shape
```

```python
x=ratings["userId"].value_counts()>50
```

```python
best_movies = x[x].index
```

```python
best_movies
```

```python
ratings=ratings[ratings["userId"].isin(best_movies)]
```

```python
ratings.shape
```

```python
ratings.head()
```

```python
movis_fulldata=ratings.merge(movies,on="movieId")
```

```python
movis_fulldata
```

```python
movis_fulldata["userId"].shape
```

```python
movis_fulldata.groupby("title")["rating"].count().reset_index()
```

```python
movie_pivot=movis_fulldata.pivot_table(columns="userId" ,index="title" , values="rating").fillna(0)
```

```python
movie_pivot
```

```python
from scipy.sparse import csr_matrix

movies_sparse= csr_matrix(movie_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors

model= NearestNeighbors(algorithm="brute"  ,metric="cosine")

model.fit(movies_sparse)
```

```python
distance , suggestion = model.kneighbors(movie_pivot.iloc[5,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
    print(movie_pivot.index[suggestion[i]])
```

```python
def movie_recommendation(movie_name):
    movie_id = np.where(movie_pivot.index==movie_name)[0][0]
    distance , suggestion = model.kneighbors(movie_pivot.iloc[movie_id,:].values.reshape(1,-1),n_neighbors=6)
    for i in range(len(suggestion)):
         print(movie_pivot.index[suggestion[i]])
    
```

```python
movie_recommendation("À nous la liberté (Freedom for Us) (1931)")
```
